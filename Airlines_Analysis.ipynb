# Analysing occupancy rate

1. For each aircraft, calculate the total revenue per year and the average revenue per ticket

```python
df=pd.read_sql_query("""select aircraft_code,ticket_count,total_revenue, total_revenue/ticket_count as Average_Revenue_per_ticket
from(select aircraft_code,count(*) as ticket_count, sum(amount) as total_revenue 
from ticket_flights join flights 
on ticket_flights.flight_id=flights.flight_id
group by aircraft_code)""",connection)
df
```

2. Caculate the average occupancy per aircraft

```python
df=pd.read_sql_query("""select aircraft_code,ticket_flights.flight_id,avg(amount) from ticket_flights inner join flights 
on ticket_flights.flight_id=flights.flight_id group by aircraft_code,fare_conditions""",connection)
df
```

```python
df=pd.read_sql_query("""select * from flights """,connection)
df
```

```python
# to find occupancy rate
occupancy_rate= pd.read_sql_query("""select a.aircraft_code,avg(a.seats_count) as booked_seats,b.num_seats,avg(a.seats_count)/b.num_seats as occupancy_rate
from
(
select aircraft_code,flights.flight_id,count(*) as seats_count 
from boarding_passes inner join flights 
on boarding_passes.flight_id=flights.flight_id
group by aircraft_code, flights.flight_id
) as a 
inner join 
(
select aircraft_code,count(*) as num_seats from seats group by aircraft_code
) as b
on a.aircraft_code=b.aircraft_code group by a.aircraft_code""",connection)

occupancy_rate
```

3.Calculate by how much the total annual turnover could increase by giving all aircraft a 10% higher occupancy rate.

```python
occupancy_rate['Increase_occupancy_rate']=occupancy_rate['occupancy_rate']*0.1
occupancy_rate
```

```python
Total_revenue=pd.read_sql_query("""select aircraft_code,sum(amount) as total_revenue from ticket_flights
join flights 
on ticket_flights.flight_id=flights.flight_id group by aircraft_code""",connection)
Total_revenue
```

```python
pd.set_option('display.float_format',str)
```

```python
# increase total annual turnover
occupancy_rate['Inc_Total_Annual_Turnover']=(Total_revenue['total_revenue']/occupancy_rate['occupancy_rate'])*occupancy_rate['Increase_occupancy_rate']
occupancy_rate
```
